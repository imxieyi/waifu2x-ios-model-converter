# Step 2: Check and download supported models
Run the cell below to get a list of supported models, then download models you want from [upscale.wiki](https://upscale.wiki/wiki/Model_Database).

```python
import yaml

import pandas as pd
from google.colab import data_table

data_table.enable_dataframe_formatter()

models = pd.DataFrame(columns=['Name', 'File', 'Scale', 'Description'])

with open('./waifu2x-ios-model-converter/known_models.yaml', 'r') as f:
    known_models_yaml = yaml.load(f.read(), Loader=yaml.Loader)
    for model in known_models_yaml['models']:
        models.loc[len(models.index)] = [model['name'], model['file'].split('/')[-1], model['scale'], model['description']] 

models
```
